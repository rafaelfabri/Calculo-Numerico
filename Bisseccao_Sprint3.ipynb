### **Como funciona o algoritimo para solução do método da Bissecção**

A primeira coisa pedida pelo sistema será a solictação da inserção da função que deseja descobrir a raiz.

* Para essa primeira parte há algumas funcionalidades importantes do sistema. Como o código foi realizado em Python, através da biblioteca matemática Numpy para representar algumas funções matemáticas é necessário chamar essa biblioteca pelo apelido dado a ela (np). Abaixo estão exemplos:

* sen(x) - representação *np.sin(x)*
* cos(x) - representação *np.cos(x)*
* tg(x) - representação *np.tan(x)*
* ln(x) - representação *np.log(x)*
* exp(x) - representação *np.exp(x)*

Para utilização de operadores matemáticos podem ser encontrados exemplos abaixo:

* 2x - 2*x
* x² - x**2
* 4√x - x**(1/4)



```python
#importando bibliotecas
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#começo do algoritimo 
print('Bem vindo ao algoritimo do Método da Bisseção\n\n')

#nessa parte pode entrar com a função que deseja encontrar a raiz
print('Entre com uma função desejada:')

equacao = input()

#atribuíndo a palavra string Repetir para a variável alternativa
alternativa = 'Repetir'

print('\nAgora você pode plotar um gráfico com o intervalo que deseja')

#laço de repetição, enquanto a variável alternativa for igual a 'Repetir' ou 'repetir' esse intervalo irá fazer o mesmo ciclo
while (alternativa == 'Repetir') or (alternativa == 'repetir'):
    print('\nentre com o intervalo que deseja ver o gráfico da função\n:')
    x1 = float(input('entre com o primeiro intervalo:'))
    x2 = float(input('\nentre com o segundo intervalo:'))

    #caso a variável x1 seja maior que x2, serão trocadas para não haver erro no código
    if x2 < x1:
        z = x1
        x1 = x2
        x2 = z

    #criando a variável y como tipo lista
    y = []
    #atribuído a variável aux um vetor de valores, esse vetor vai de x1 até (x2-0.01) pulando de 0.01 a 0.01
    aux = np.arange(x1, x2, 0.01) 
    
    #laço de repetição onde irá atrubuir cada indice do vetor aux para x, quando atribuído será cálculado o valor de y da equação
    #for i in range(indice 0 do vetor, len(aux)=tamanho do vetor, 1)
    for i in range(0, len(aux), 1):
        x = float(aux[i])
        y.append(eval(equacao))

    #plotar gráfico de linha
    fig, ax = plt.subplots(figsize=(8,6))
    sns.lineplot(x=list(aux), y=y, ax=ax)
    fig.savefig("mygraph.png")

    print('\n\nVocê encontrou alguma raiz com intervalo verificado?\n')

    print('Caso não tenha encontrado, digite repetir para colocar outro intervalo para ver o gráfico\n')

    print('Caso tenha encontrado apenas digite enter\n')

    #atualizar variável alternativa
    alternativa = input('Digite aqui...');



resultado = 10

#criando novo laço de repetição, que enquanto resultado for maior que 0 continuará a se repetir
while resultado > 0:
    print('\n\nDigite abaixo o intervalo para encontrar a raiz pelo método da Bissecção\n')
    x_1 = float(input('entre com o menor valor do intervalo'))
    x_2 = float(input('entre com o maior valor do intervalo'))

    #atribuíndo para x o valor da variável x_1 e depois calculando o valor de f(x_1) e atribuíndo para aux_1
    x = float(x_1)
    aux_1 = eval(equacao)

    #atribuíndo para o valor de x para a variável x_2 e depois calculando f(x_2) e multiplicando com f(x_1).
    #Portanto, a variável resultado é atualizada com o valor do Teorema de Bozano
    x = float(x_2)
    resultado = aux_1*eval(equacao)

    #condição caso o teorema de Bozano seja menor que 0 ali se tem uma raiz no intervalo - e poderá entrar com o valor do critério de parada
    if resultado<0:
        print('Teorema de Bozano');
        print('f(a)*f(b) < 0 :', round(resultado,4), '< 0');
        print('\n');
        print('Parábens, há uma raiz neste intervalo!\n');
        min_error = float(input('Entre com o valor para o critério de parada:'));
     
    #caso o valor seja maior que irá imprimir as coisas abaixo e continuar repetindo o laço 
    else:
        print('f(a)*f(b) < 0 :', resultado, '> 0');
        print('Por favor, repita o intervalo, pois não há nenhuma raiz no intervalo\n')

calculo_erro = []
x_medio_lista = []

x_medio = float((x_1 + x_2) / 2)
x_medio_lista.append(x_medio)

erro = 1000.0
calculo_erro.append(0)

i=0

#criando um novo laço de repetição, enquanto o erro calculado for maior que a precisão deseja fornecida pelo usuário o laço continuará a se repetir
while (erro > min_error):
    
    #obtendo o valor do teorema de bozano
    x = float(x_1)
    resultado_x_1 = eval(equacao)
    x = float(x_medio)
    resultado_x_medio = eval(equacao)
    
    #atualizando o valor da iteração
    i = i + 1

    #caso resultado  de f(a)*f(xi) seja menor que 0 fazer o que está abaixo
    if (resultado_x_1*resultado_x_medio) < 0:
        aux_x_medio = float(x_medio)
        x_2 = float(x_medio) 
        x_medio = float((x_1 + x_2) / 2 )
        x_medio_lista.append(x_medio)
        erro = ( abs(x_medio - aux_x_medio)/x_medio )
        calculo_erro.append(erro)

    #caso o resultado de f(a)*f(xi) seja maior que 0 fazer o que está abaixo
    elif (resultado_x_1*resultado_x_medio) > 0:
        aux_x_medio = float(x_medio)
        x_1 = float(x_medio)
        x_medio = float((x_1 + x_2) / 2) 
        x_medio_lista.append(x_medio)
        erro = abs((x_medio - aux_x_medio) / x_medio ) 
        calculo_erro.append(erro)

    #caso o resultado seja f(a)*f(xi) = 0 fazer o que está abaixo
    elif (resultado_x_1*resultado_x_medio) == 0:
        erro = 0
        calculo_erro.append(erro)

print('\n')

print('x médio', '   ','erro')
for i in range(0, len(x_medio_lista), 1):
    print(round(x_medio_lista[i], 4), '   ', round(calculo_erro[i], 4))

print('A raiz da função é:', round(x_medio,4))
```
